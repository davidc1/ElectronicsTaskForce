```python
def getFTvec(run,cr,sl,ev,pl):
    chL = []
    if (pl == 'U'):
        chL = [2*x for x in xrange(16)]
    if (pl == 'V'):
        chL = [2*x+1 for x in xrange(16)]
    if (pl == 'Y'):
        [32+x for x in xrange(32)]

    wf_avg = np.zeros(9595)

    for x in chL:
        if (run == 380):
            wf_avg += np.array(adc_dict_380[(cr,sl,x,ev)])
        if (run == 379):
            wf_avg += np.array(adc_dict_379[(cr,sl,x,ev)])
        if (run == 378):
            wf_avg += np.array(adc_dict_378[(cr,sl,x,ev)])
        if (run == 377):
            wf_avg += np.array(adc_dict_377[(cr,sl,x,ev)])
    wf_avg /= 16.
    
    fftvec = ft.rfft(wf_avg)
    times  = ft.rfftfreq(len(wf_avg),0.5E-3)
    return fftvec, times

def getFTvecSingleChan(run,cr,sl,ch,ev):


    wf_avg = np.zeros(9595)

    if (run == 380):
        wf_avg = np.array(adc_dict_380[(cr,sl,ch,ev)])
    if (run == 379):
        wf_avg = np.array(adc_dict_379[(cr,sl,ch,ev)])
    if (run == 378):
        wf_avg = np.array(adc_dict_378[(cr,sl,ch,ev)])
    if (run == 377):
        wf_avg = np.array(adc_dict_377[(cr,sl,ch,ev)])
    
    fftvec = ft.rfft(wf_avg)
    times  = ft.rfftfreq(len(wf_avg),0.5E-3)
    return fftvec, times
```
