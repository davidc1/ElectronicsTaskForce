```python
def getFTvec(run,cr,sl,ev,pl):
    chL = []
    if (pl == 'U'):
        chL = [2*x for x in xrange(16)]
    if (pl == 'V'):
        chL = [2*x+1 for x in xrange(16)]
    if (pl == 'Y'):
        [32+x for x in xrange(32)]

    wf_avg = np.zeros(9595)

    for x in chL:
        if (run == 380):
            wf_avg += np.array(adc_dict_380[(cr,sl,x,ev)])
        if (run == 379):
            wf_avg += np.array(adc_dict_379[(cr,sl,x,ev)])
        if (run == 378):
            wf_avg += np.array(adc_dict_378[(cr,sl,x,ev)])
        if (run == 377):
            wf_avg += np.array(adc_dict_377[(cr,sl,x,ev)])
    wf_avg /= 16.
    
    fftvec = ft.rfft(wf_avg)
    times  = ft.rfftfreq(len(wf_avg),0.5E-3)
    return fftvec, times
```

```python
avg_fft_380 = np.zeros(9595/2+1)
avg_fft_379 = np.zeros(9595/2+1)
avg_fft_378 = np.zeros(9595/2+1)
avg_fft_377 = np.zeros(9595/2+1)

for n in xrange(100):
    fftvec,times = getFTvec(380,4,11,n,'V')
    avg_fft_380 += np.absolute(np.array(fftvec))
    fftvec,times = getFTvec(379,4,11,n,'V')
    avg_fft_379 += np.absolute(np.array(fftvec))
    fftvec,times = getFTvec(378,4,11,n,'V')
    avg_fft_378 += np.absolute(np.array(fftvec))
    fftvec,times = getFTvec(377,4,11,n,'V')
    avg_fft_377 += np.absolute(np.array(fftvec))
    
fig = plt.figure(figsize=(15,6))
plt.plot(times,avg_fft_380/100.,'r',label='1.0 usec')
plt.plot(times,avg_fft_379/100.,'b',label='0.5 usec')
plt.plot(times,avg_fft_378/100.,'g',label='3.0 usec')
plt.plot(times,avg_fft_377/100.,'m',label='2.0 usec')
plt.grid()
plt.legend(loc=2)
plt.ylabel('FFT Mag. Amplitude')
plt.xlabel('Freq. kHz')
plt.ylim([0,600])
plt.xlim([800,1000])
plt.title('Fourier Transform for Channels [%i,%i,All %s]'%(4,11,'V'))
plt.savefig('../ShapingTime/comparison_ZOOM.png')
plt.show()
```
