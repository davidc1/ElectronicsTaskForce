```python
import ROOT
import sys, os
import numpy as np
import numpy.fft as ft
from root_numpy import root2array, root2rec, tree2rec, list_branches
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline


plt.rcParams.update({'font.size': 16})

run = 307


fname = "run_%i_tree.root"%(run)
print fname
f = ROOT.TFile(fname)

ch = f.Get("_wf_tree")
ch_arr = tree2rec(ch,branches=['crate','fem','chan','rms','mean','adc_v','event'])
```

```python
import pandas as pd
df = pd.DataFrame(ch_arr)
```

```python
adc_dict = {}
for y in xrange(len(ch_arr)):
    adc_dict[(ch_arr['crate'][y],ch_arr['fem'][y],ch_arr['chan'][y],ch_arr['event'][y])] = ch_arr['adc_v'][y]-ch_arr['mean'][y]
```

```python
rms = df.query('chan>32 and crate>1 and crate <9')['rms'].ravel()
plt.hist(rms,np.linspace(0,3,100),log=False)
plt.show()
```

```python
adcs = df.query('crate==4 and fem==11 and chan==15 and event==0')['adc_v'].ravel()
```

```python
fig = plt.figure(figsize=(15,6))
plt.plot(adcs[0][600:800])
plt.show()
```

```python
import numpy.fft as fft

cr = 4
sl = 11
ch = 0
ev = 1

chU = [2*x for x in xrange(16)]
chV = [2*x+1 for x in xrange(16)]
chY = [32+x for x in xrange(32)]


pl = 'U'

wf_avg = np.zeros(9595)

for x in chU:
    wf_avg += np.array(adc_dict[(cr,sl,x,ev)])
wf_avg /= 16.

fftvec = ft.rfft(wf_avg)
times  = ft.rfftfreq(len(wf_avg),0.5E-3)

#fftvec = ft.rfft(adc_dict[(cr,sl,ch,ev)])
#times  = ft.rfftfreq(len(adc_dict[(cr,sl,ch,ev)]),0.5E-3)

fig = plt.figure(figsize=(15,6))
plt.plot(wf_avg,color='b')
plt.xlim([0,9595])
plt.grid()
plt.ylabel('ADCs')
plt.xlabel('Time Samples')
plt.title('Waveform for Run %i Channel [%i,%i,All %s] Event %i'%(run,cr,sl,pl,ev))
plt.savefig('wf_run_%i_cr_%i_sl_%i_ch_all%s_ev_%i.png'%(run,cr,sl,pl,ev))
plt.show()

fig = plt.figure(figsize=(15,6))
plt.plot(times,np.absolute(fftvec),color='r')
plt.grid()
plt.ylabel('FFT Mag. Amplitude')
plt.xlabel('Freq. kHz')
plt.title('Fourier Transform Run %i Channel [%i,%i,All %s] Event %i'%(run,cr,sl,pl,ev))
plt.savefig('fft_run_%i_cr_%i_sl_%i_ch_all%s_ev_%i.png'%(run,cr,sl,pl,ev))
plt.show()
```
