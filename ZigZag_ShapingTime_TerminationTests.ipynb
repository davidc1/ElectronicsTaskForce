```python
import ROOT
import sys, os
import numpy as np
import numpy.fft as ft
from root_numpy import root2array, root2rec, tree2rec, list_branches
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline


plt.rcParams.update({'font.size': 16})

nevents = 10

query = 'event<=%i'%nevents
print query

crate = 4
fem = 12

#runs = [380,482,502,512,522,532,538]
#runs = [440,460,476,499,522]
#runs = [532,533,534,535,536,537,538]
runs = [522,523]
arr_v = []

for r in runs:

    fname = '../elec_test_%i_tree.root'%r
    f     = ROOT.TFile(fname)
    tree  = f.Get("_wf_tree")
    arr_v.append(tree2rec(tree,branches=['crate','fem','chan','rms','mean','adc_v','event'],selection=query))
```

```python
# vector linking (crate,fem,chan) -> element in array
adc_dict_v = []

for r in xrange(len(runs)):
    adc_dict = {}
    for y in xrange(len(arr_v[r])):
        adc_dict[(arr_v[r]['crate'][y],arr_v[r]['fem'][y],arr_v[r]['chan'][y],arr_v[r]['event'][y])] = y#arr_v[r]['adc_v'][y]-arr_v[r]['mean'][y]
    adc_dict_v.append(adc_dict)
```

```python
def getFTvec(run,cr,sl,ev,pl):
    chL = []
    if (pl == 'U'):
        chL = [2*x for x in xrange(16)]
    if (pl == 'V'):
        chL = [2*x+1 for x in xrange(16)]
    if (pl == 'Y'):
        [32+x for x in xrange(32)]

    wf_avg = np.zeros(9595)

    for x in chL:
        wf_avg += np.array(arr_v[run]['adc_v'][adc_dict_v[run][(cr,sl,x,ev)]]-\
                           arr_v[run]['mean'][adc_dict_v[run][(cr,sl,x,ev)]])

    wf_avg /= 16.
    
    fftvec = ft.rfft(wf_avg)
    times  = ft.rfftfreq(len(wf_avg),0.5E-3)
    return fftvec, times

def getFTvecSingleChan(run,cr,sl,ch,ev):


    wf_avg = np.zeros(9595)

    wf_avg = np.array(arr_v[run]['adc_v'][adc_dict_v[run][(cr,sl,ch,ev)]]-\
                      arr_v[run]['mean'][adc_dict_v[run][(cr,sl,ch,ev)]])

    fftvec = ft.rfft(wf_avg)
    times  = ft.rfftfreq(len(wf_avg),0.5E-3)
    return fftvec, times
```
